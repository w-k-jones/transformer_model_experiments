```python
class LitProgressBar(TQDMProgressBar):
    def init_validation_tqdm(self):
        bar = tqdm(            
            disable=True,            
        )
        return bar

# Add a logger for metrics and checkpoints
csv_logger = CSVLogger(
    save_dir='',
    name='logs'
)
```

```python
class GeoCloudsat(LightningModule):
    def __init__(self, *args, **kwargs):
        super().__init__()
        self.model = GeoCloudSatTransformer(*args, **kwargs)

    def forward(self, batch):
        B, C, H, W = batch[1].shape
        batch[1] = torch.cat([
            batch[1], 
            torch.zeros(B, 1, H, W, dtype=batch[1].dtype, device=batch[1].device),
        ], dim=1,
        )
        B, C, L = batch[3].shape
        batch[3] = torch.cat([
            batch[3], 
            torch.zeros(B, 1, L, dtype=batch[3].dtype, device=batch[3].device) + 1,
        ], dim=1,
        )
        return self.model.forward(*batch)

    def training_step(self, batch, batch_idx):
        B, C, H, W = batch[1].shape
        batch[1] = torch.cat([
            batch[1], 
            torch.zeros(B, 1, H, W, dtype=batch[1].dtype, device=batch[1].device),
        ], dim=1,
        )
        B, C, L = batch[3].shape
        batch[3] = torch.cat([
            batch[3], 
            torch.zeros(B, 1, L, dtype=batch[3].dtype, device=batch[3].device) + 1,
        ], dim=1,
        )
        loss = self.model.forward_loss(*batch)
        # log training loss
        self.log(
            "train_loss", loss, on_step=True, on_epoch=True, logger=True, prog_bar=True
        )
        return loss
    
    def validation_step(self, batch, batch_idx):
        B, C, H, W = batch[1].shape
        batch[1] = torch.cat([
            batch[1], 
            torch.zeros(B, 1, H, W, dtype=batch[1].dtype, device=batch[1].device),
        ], dim=1,
        )
        B, C, L = batch[3].shape
        batch[3] = torch.cat([
            batch[3], 
            torch.zeros(B, 1, L, dtype=batch[3].dtype, device=batch[3].device) + 1,
        ], dim=1,
        )
        loss = self.model.forward_loss(*batch)
        # log validation loss
        self.log(
            "val_loss", loss, on_step=True, on_epoch=True, logger=True, prog_bar=True
        )
        return loss

    def test_step(self, batch, batch_idx):
        B, C, H, W = batch[1].shape
        batch[1] = torch.cat([
            batch[1], 
            torch.zeros(B, 1, H, W, dtype=batch[1].dtype, device=batch[1].device),
        ], dim=1,
        )
        B, C, L = batch[3].shape
        batch[3] = torch.cat([
            batch[3], 
            torch.zeros(B, 1, L, dtype=batch[3].dtype, device=batch[3].device) + 1,
        ], dim=1,
        )
        loss = self.model.forward_loss(*batch)
        # log test loss
        self.log(
            "test_loss", loss, on_step=True, on_epoch=True, logger=True, prog_bar=True
        )
        return loss
        
    def configure_optimizers(self):
        optimizer = torch.optim.AdamW(self.parameters(), lr=5e-4)

        return {
            "optimizer": optimizer,
            "lr_scheduler": {"scheduler": ExponentialLR(optimizer, gamma=0.95)}
            # "lr_scheduler": {
            #     "scheduler": ReduceLROnPlateau(optimizer),
            #     "patience": 10,
            #     "interval": "epoch",
            #     "monitor": "val_loss",
            #     "frequency": 1,
            # },
        }
```

```python
_ = seed_everything(42)

dl = GeoCloudsatDataloader(16, files)

model = GeoCloudsat(
    in_chans=16,
    img_size=256, 
    patch_size=16, 
    embed_dim=576, 
    depth=8,
    num_heads=8,
    decoder_embed_dim=576,
    decoder_depth=8, 
    decoder_num_heads=8,
    mlp_ratio=4.,
    num_coords=3,
    coord_limits=[
        [1  / (2*np.pi), 360 / (2*np.pi)], # longitude 1->360 degrees
        [1  / (2*np.pi), 360 / (2*np.pi)], # latitude 1->360 degrees
        [2, 2*np.pi], 
    ],
)

# model = torch.compile(model)
```

```python
# Initialize the trainer
trainer = Trainer(
    max_epochs=20,
    callbacks=[LitProgressBar()],
    logger=csv_logger,
    accelerator="auto",
    accumulate_grad_batches=1,
)

trainer.fit(model, dl)
```
